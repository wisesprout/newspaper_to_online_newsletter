```python
import time
start = time.time()  # 셀 실행시간 재기


#ray.init()  # Ray 실행

@ray.remote  # Ray Task(Remote Function) 정의, 기존 함수에 @ray.remote만 추가
def split_sentences(corpus):
    splited_sentences = []

    for sentences in corpus:
        sentences = sentences.replace("'", "")
        sentences = sentences.replace("\"", "")
        sentences = sentences.split("\n")

        for sentence in sentences:
            splited_sentences.extend(kss.split_sentences(sentence))
    return splited_sentences

def chunker_list(seq, size):
    return (seq[i::size] for i in range(size))

# 프로세스 개수
process_num = 30

# 프로세스 개수만큼 news 분리
sentences_chunk = list(chunker_list(dataset, process_num))

# ray를 이용해 멀티프로세싱
# 함수를 일반 python 함수처럼 실행하면 에러 발생
# 함수명.remote()로 task 호출
# task를 동시에 process_num번 실행하기
futures = [split_sentences.remote(sentences_chunk[i]) for i in range(process_num)]

# 결과를 1-d로 합치기
results = ray.get(futures)  # 값 받기
results = list(chain.from_iterable(results))


end = time.time()
print('전처리 시간:',end-start)
```

- pytorch custom dataset 생성
    - 띄어쓰기에 대한 표현
    - 한글 문장을 BERT의 input feature로 생성


```python
class CorpusDataset(Dataset):
    def __init__(self, data_path: str, transform: Callable[[List, List], Tuple]):  # input feature을 만드는 transform function을 넘겨받아 __getitem__()에 내용 추가
        self.sentences = []
        self.slot_labels = ["UNK", "PAD", "B", "I"]
        self.transform = transform

        self._load_data(data_path)


    def _load_data(self, data_path: str):
        """data를 file에서 불러온다.

        Args:
            data_path: file 경로
        """
        with open(data_path, mode="r", encoding="utf-8") as f:  # with문 끝나면 자동 .close
            lines = f.readlines()  # 각 줄을 요소로 갖는 리스트 lines
            self.sentences = [line.split() for line in lines]  # lines의 요소를 단어별로 분리한 리스트 sentences

    def _get_tags(self, sentence: List[str]) -> List[str]:
        """문장에 대해 띄어쓰기 tagging을 한다.
        character 단위로 분리하여 BI tagging을 한다.

        Args:
            sentence: 문장

        Retrns:
            문장의 각 토큰에 대해 tagging한 결과 리턴
            ["B", "I"]
        """

        tags = []
        for word in sentence:  # sentences의 요소별 = 한 단어
            for i in range(len(word)):  # 단어의 길이만큼 반복
                if i == 0:
                    tags.append("B")  # 단어의 첫글자 태그
                else:
                    tags.append("I")  # 단어의 나머지 글자 태그
        return tags  # 글자의 태그를 요소로 갖는 리스트 tags

    def __len__(self):
        '''한 단어의 길이 반환'''
        return len(self.sentences)

    def __getitem__(self, idx):
        sentence = "".join(self.sentences[idx])
        sentence = [s for s in sentence]
        tags = self._get_tags(self.sentences[idx])
        tags = [self.slot_labels.index(t) for t in tags]

        (
            input_ids,
            attention_mask,
            token_type_ids,
            slot_label_ids, 
        ) = self.transform(sentence, tags)

        return input_ids, attention_mask, token_type_ids, slot_label_ids
```

- transform function 구현
- pretrained KoBERT 사용

```python
class Preprocessor:
    def __init__(self, max_len: int):
        self.tokenizer = KoBertTokenizer.from_pretrained("monologg/kobert")
        self.max_len = max_len
        self.pad_token_id = 0

    def get_input_features(
        self, sentence: List[str], tags: List[str]
    ) -> Tuple[torch.Tensor, torch.Tensor, torch.Tensor, torch.Tensor]:
        """문장과 띄어쓰기 tagging에 대해 feature로 변환한다.

        Args:
            sentence: 문장
            tags: 띄어쓰기 tagging

        Returns:
            feature를 리턴한다.
            input_ids, attention_mask, token_type_ids, slot_labels
        """

        input_tokens = []
        slot_label_ids = []
					
        # tokenize
        for word, tag in zip(sentence, tags):
            tokens = self.tokenizer.tokenize(word)

            if len(tokens) == 0:
                tokens = self.tokenizer.unk_token

            input_tokens.extend(tokens)

            for i in range(len(tokens)):
                if i == 0:
                    slot_label_ids.extend([tag])
                else:
                    slot_label_ids.extend([self.pad_token_id])

        # max_len보다 길이가 길면 뒤에 자르기
        if len(input_tokens) > self.max_len - 2:
            input_tokens = input_tokens[: self.max_len - 2]
            slot_label_ids = slot_label_ids[: self.max_len - 2]

        # cls, sep 추가
        input_tokens = (
            [self.tokenizer.cls_token] + input_tokens + [self.tokenizer.sep_token]
        )
        slot_label_ids = [self.pad_token_id] + slot_label_ids + [self.pad_token_id]

        # token을 id로 변환
        input_ids = self.tokenizer.convert_tokens_to_ids(input_tokens)

        attention_mask = [1] * len(input_ids)
        token_type_ids = [0] * len(input_ids)

        # padding
        pad_len = self.max_len - len(input_tokens)
        input_ids = input_ids + ([self.tokenizer.pad_token_id] * pad_len)
        slot_label_ids = slot_label_ids + ([self.pad_token_id] * pad_len)
        attention_mask = attention_mask + ([0] * pad_len)
        token_type_ids = token_type_ids + ([0] * pad_len)

        input_ids = torch.tensor(input_ids, dtype=torch.long)
        attention_mask = torch.tensor(attention_mask, dtype=torch.long)
        token_type_ids = torch.tensor(token_type_ids, dtype=torch.long)
        slot_label_ids = torch.tensor(slot_label_ids, dtype=torch.long)

        return input_ids, attention_mask, token_type_ids, slot_label_ids
```

# 3. CREATE MODEL

- set hyperparameters

```python
# configuration

task = 'korean_spacing_20220316'
log_path = root+'spacing_logs'
bert_model = 'monologg/kobert'
train_data_path = root+'spacing_datasets/train.txt'
val_data_path = root+'spacing_datasets/val.txt'
test_data_path = root+'spacing_datasets/test.txt'
max_len = 128
train_batch_size = 128
eval_batch_size = 128
dropout_rate = 0.1
gpus = 1  # gpu 없으면 0, 있으면 8
distributed_backend = 'dp'  # Distributed Data Parallel=gpu 활용 옵션
```

- create model using PyTorch Lightning

```python
class SpacingBertModel(pl.LightningModule):
    def __init__(self, dataset: CorpusDataset):
        super().__init__()
        #self.config = config
        self.dataset = dataset

        # define model
        self.slot_labels_type = ["UNK", "PAD", "B", "I"]
        self.pad_token_id = 0
        self.bert_config = BertConfig.from_pretrained(
            bert_model, num_labels = len(self.slot_labels_type)
        )
        self.model = BertModel.from_pretrained(
            bert_model, config=self.bert_config
        )
        self.dropout = nn.Dropout(dropout_rate)
        self.linear = nn.Linear(
            self.bert_config.hidden_size, len(self.slot_labels_type)
        )


    # forward
    def forward(self, input_ids, attention_mask, token_type_ids):
        outputs = self.model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids,
        )

        x = outputs[0]
        x = self.dropout(x)
        x = self.linear(x)

        return x


    '''results from training'''
    def training_step(self, batch, batch_nb):

        input_ids, attention_mask, token_type_ids, slot_label_ids = batch
            
        outputs = self(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids,
        )
        
        loss = self._calculate_loss(outputs, slot_label_ids)
        tensorboard_logs = {"train_loss": loss}
        
        return {"loss": loss, "log": tensorboard_logs}

    def _calculate_loss(self, outputs, labels):
        active_logits = outputs.view(-1, len(self.slot_labels_type))
        active_labels = labels.view(-1)
        loss = F.cross_entropy(active_logits, active_labels)

        return loss

    def _f1_score(self, gt_slot_labels, pred_slot_labels):
        return torch.tensor(
            f1_score(gt_slot_labels, pred_slot_labels), dtype=torch.float32
        )

    # accuracy score 추가
    def _acc_score(self, gt_slot_labels, pred_slot_labels):
        return torch.tensor(
            accuracy_score(gt_slot_labels, pred_slot_labels), dtype=torch.float32
        )
    
    # precision score 추가
    def _precision_score(self, gt_slot_labels, pred_slot_labels):
        return torch.tensor(
            precision_score(gt_slot_labels, pred_slot_labels), dtype=torch.float32
        )

    # recall score 추가
    def _recall_score(self, gt_slot_labels, pred_slot_labels):
        return torch.tensor(
            recall_score(gt_slot_labels, pred_slot_labels), dtype=torch.float32
        )

    # classificaion_report score 추가
    def _report(self, gt_slot_labels, pred_slot_labels):
        return torch.tensor(
            classification_report(gt_slot_labels, pred_slot_labels), dtype=torch.float32
        )


    '''perform validation at every epoch'''
    def validation_step(self, batch, batch_nb):
        input_ids, attention_mask, token_type_ids, slot_label_ids = batch

        outputs = self(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids,
        )

        loss = self._calculate_loss(outputs, slot_label_ids)
        pred_slot_labels, gt_slot_labels = self._convert_ids_to_labels(
            outputs, slot_label_ids
        )

        val_f1 = self._f1_score(gt_slot_labels, pred_slot_labels)
        # 추가
        self.log('val_loss', torch.tensor([loss]))
        val_acc = self._acc_score(gt_slot_labels, pred_slot_labels)
        val_precision = self._precision_score(gt_slot_labels, pred_slot_labels)
        val_recall = self._recall_score(gt_slot_labels, pred_slot_labels)

        return {"val_loss": loss, "val_f1": val_f1, 'val_acc': val_acc, 'val_precision': val_precision, 'val_recall' : val_recall}

    # validation이 끝날 때마다 하고 싶은 것은 여기에 추가
    def validation_epoch_end(self, outputs):
        val_loss = torch.stack([x["val_loss"] for x in outputs]).mean()
        val_f1 = torch.stack([x["val_f1"] for x in outputs]).mean()
        val_acc = torch.stack([x["val_acc"] for x in outputs]).mean()
        val_precision = torch.stack([x["val_precision"] for x in outputs]).mean()
        val_recall = torch.stack([x["val_recall"] for x in outputs]).mean()
    
        tensorboard_log = {
            "val_loss": val_loss,
            "val_f1": val_f1,
            "val_acc": val_acc,
            "val_precision": val_precision,
            "val_recall": val_recall
        }

        return {"val_loss": val_loss, "progress_bar": tensorboard_log} 
    

    '''학습이 끝난 후 평가할 test step'''
    def test_step(self, batch, batch_nb):
        input_ids, attention_mask, token_type_ids, slot_label_ids = batch

        outputs = self(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids,
        )

        pred_slot_labels, gt_slot_labels = self._convert_ids_to_labels(
            outputs, slot_label_ids
        )

        test_f1 = self._f1_score(gt_slot_labels, pred_slot_labels)
        test_acc = self._acc_score(gt_slot_labels, pred_slot_labels)
        test_precision = self._precision_score(gt_slot_labels, pred_slot_labels)
        test_recall = self._recall_score(gt_slot_labels, pred_slot_labels)

        test_step_outputs = {
            "test_f1": test_f1,
            "test_acc": test_acc,
            "test_precision": test_precision,
            "test_recall": test_recall
        }
    
        return test_step_outputs

    # 실제 추론한 결과를 리턴 받고 싶으면 test_epoch_end()의 return 값에 추가
    def test_epoch_end(self, outputs):
        test_f1 = torch.stack([x["test_f1"] for x in outputs]).mean()
        test_acc = torch.stack([x["test_acc"] for x in outputs]).mean()
        test_precision = torch.stack([x["test_precision"] for x in outputs]).mean()
        test_recall = torch.stack([x["test_recall"] for x in outputs]).mean()

        test_step_outputs = {
            "test_f1": test_f1,
            "test_acc": test_acc,
            "test_precision": test_precision,
            "test_recall": test_recall
        }

        return test_step_outputs


    # config optimizer
    def configure_optimizers(self):
        return AdamW(self.model.parameters(), lr=2e-5, eps=1e-8)


    # dataloader
    def train_dataloader(self):
        return DataLoader(self.dataset["train"], batch_size=train_batch_size)

    def val_dataloader(self):
        return DataLoader(self.dataset["val"], batch_size=eval_batch_size)

    def test_dataloader(self):
        return DataLoader(self.dataset["test"], batch_size=eval_batch_size)


    def _convert_ids_to_labels(self, outputs, slot_labels):
        _, y_hat = torch.max(outputs, dim=2)
        y_hat = y_hat.detach().cpu().numpy()
        slot_label_ids = slot_labels.detach().cpu().numpy()

        slot_label_map = {i: label for i, label in enumerate(self.slot_labels_type)}
        slot_gt_labels = [[] for _ in range(slot_label_ids.shape[0])]
        slot_pred_labels = [[] for _ in range(slot_label_ids.shape[0])]

        for i in range(slot_label_ids.shape[0]):
            for j in range(slot_label_ids.shape[1]):
                if slot_label_ids[i, j] != self.pad_token_id:
                    slot_gt_labels[i].append(slot_label_map[slot_label_ids[i][j]])
                    slot_pred_labels[i].append(slot_label_map[y_hat[i][j]])

        return slot_pred_labels, slot_gt_labels

```

#4. TRAIN MODEL

- get dataset and model

```python
dataset = {}
dataset["train"] = CorpusDataset(train_data_path, preprocessor.get_input_features)
dataset["val"] = CorpusDataset(val_data_path, preprocessor.get_input_features)
dataset["test"] = CorpusDataset(test_data_path, preprocessor.get_input_features)

bert_finetuner = SpacingBertModel(dataset)
```

- logging과 학습에 사용할 callback 작성
    - checkpoint_callback : 매 epoch마다 validation loss를 계산해 loss가 줄어드는 경우 checkpoint를 저장
    - early_stop_callback : validation loss가 더 이상 줄어들지 않으면 학습을 종료

```python
logger = TensorBoardLogger(
    save_dir=os.path.join(log_path, task), version=1, name=task
)

checkpoint_callback = ModelCheckpoint(
    filepath=root+"checkpoints/"+ task + "/{epoch}_{val_loss:35f}",
    verbose=True,
    monitor="val_loss",
    mode="min",
    save_top_k=3,
    prefix="",
)

early_stop_callback = EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=3,
    verbose=False,
    mode="min",
)
```

- 학습 수행

```python
trainer = pl.Trainer(
    gpus=gpus,
    #strategy=distributed_backend,
    #strategy='ddp',
    checkpoint_callback=checkpoint_callback,
    callbacks=[early_stop_callback],
    logger=logger
)

trainer.fit(bert_finetuner)
trainer.test()
```
